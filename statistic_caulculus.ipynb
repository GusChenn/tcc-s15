```python
from scipy.stats import binom
import matplotlib.pyplot as plt

# Approximated parameters
n = 60
p = 0.7681
x_values = list(range(n + 1))

def sum(start, end, probabilities):
    accumulated_value = 0
    range_to_ve_summed = range(start, end)
    for x in range_to_ve_summed:
        accumulated_value += probabilities[x]
    return accumulated_value

mean, var = binom.stats(n, p)

dist = [binom.pmf(x, n, p) for x in x_values]

print("x\tP(X = x)")
for i in range(n + 1):
    print(str(x_values[i]) + "\t" + str(dist[i]))

print("mean = " + str(mean))
print("variance = " + str(var))

plt.bar(x_values, dist)
plt.show()

print(sum(30, 60, dist))
print(sum(0, 30, dist))
print(sum(0, 60, dist))
```
